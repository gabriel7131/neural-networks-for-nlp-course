```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.random((100000, 2)) * 4 - (2, 0)
y = np.asarray([0 if x[1] < x[0]**2 else 1 for x in X])
c = ['r' if l else 'b' for l in y]
plt.scatter(X[:, 0], X[:, 1], marker='s', c=c)
```
